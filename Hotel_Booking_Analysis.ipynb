# Data Analysis and Visualization

```python
hotel_perc = df['hotel'].value_counts(normalize = True)
hotel_count = df['hotel'].value_counts()
print(hotel_count,'\n','-'*30,hotel_perc,'\n')

sns.set_style('dark')

sns.countplot(x='hotel',data=df)
```

```python
canceled_count = df['is_canceled'].value_counts()
canceled_perc = df['is_canceled'].value_counts(normalize = True)
print(canceled_count,'\n','-'*30,canceled_perc,'\n')

plt.figure(figsize = (5,4))
plt.title('Cancelation Status')
plt.bar(['Notcanceled','Canceled'],df['is_canceled'].value_counts())
plt.show()
```

```python
plt.figure(figsize=(5,4))
sns.countplot(x='hotel',data=df,hue='is_canceled')

plt.title('Cancelation status in different hotels',size=15)
plt.xlabel('Hotels')
plt.ylabel('No. of Reservations')
```
